```python
# IF CSV HAS MIXED COLUMNS

import csv

with open('tabula-Tabla_casos_positivos_COVID-19_resultado_InDRE_2020.03.31.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        if(row[0][0:2] == 'NA'):
            #print('c1')
            l1 = row[0].replace(" ", ",")
            l1.split(',')
            print(l1[3:] + ',' + row[1] + ',' + row[2] + ',' + row[3] + ',' + row[4] + ',' + l1[0:2])
        else:
            if(len(row[0]) > 3):
                if(row[0][2] == '/'):
                    #print('c2')
                    l1 = row[0].replace(" ", ",")
                    l1.split(',')
```

```python
# CLEANING DATA

f = open("tabula-Tabla_casos_positivos_COVID-19_resultado_InDRE_2020.04.03.csv", "r", encoding='utf-8')
for x in f:
    y = x
    y = y.replace("CIUDAD,DE,MEXICO", "CIUDAD DE MEXICO")
    y = y.replace("BAJA,CALIFORNIA", "BAJA CALIFORNIA")
    y = y.replace("BAJA,CALIFORNIA,SUR", "BAJA CALIFORNIA SUR")
    y = y.replace("CALIFORNIA,SUR", "CALIFORNIA SUR")
    y = y.replace("QUINTANA,ROO", "QUINTANA ROO")
    y = y.replace("NUEVO,LEON", "NUEVO LEON")
    y = y.replace("SAN,LUIS,POTOSI", "SAN LUIS POTOSI")  
    y = y.replace("", "n")
    y = y.replace("", "n")
    y = y.replace("", "A")
    y = y.replace("", "E")
    y = y.replace("", "I")
    y = y.replace("", "O")
    y = y.replace("", "U")
    y = y.replace("", "n")
    y = y.replace("", "N")
    y = y.replace("", "n")
    y = y.rstrip('\n')
    print(y)
```

```python
# MERGE LAT AND LONG

import math
import pandas as pd

data_confirmed        = pd.read_csv('tabula-Tabla_casos_positivos_COVID-19_resultado_InDRE_2020.04.03.csv')
data_lat_long_estados = pd.read_csv('estados_lat_long.csv')

data_confirmed['LONG'] = 0.0
data_confirmed['LAT'] = 0.0

for i in data_confirmed['ESTADO'].unique():
    counter = 0
    for index, row in data_confirmed.iterrows():
        if row['ESTADO'] == i:
            counter = counter + 1
    print(str(i) + ": " + str(counter))    

for index, row in data_confirmed.iterrows():
    for ind,r in data_lat_long_estados.iterrows():
        if row['ESTADO']==r['ESTADO']:
            data_confirmed.set_value(index,'LAT',r['LAT'])
            data_confirmed.set_value(index,'LONG',r['LONG'])
```
